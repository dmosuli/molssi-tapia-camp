```python
import os
```

```python
import numpy
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)
print(headers,data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(data[:,1])
```

```python
sample = headers[1]
sample2 = headers[2]
sample3 = headers[3]


plt.plot(data[:,1], label=sample)
plt.plot(data[:,2], label=sample2)
plt.plot(data[:,3], label=sample3)
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.legend()
plt.savefig('twosamples.png',dpi=300)  # dpi for resolution
```

```python
num_columns = len(headers)
print(headers)
#print(F'the number of columns is {num_columns}')

for col in range(1,num_columns):
    plt.figure()
    #print(F"the column number is {col}.")
    plt.plot(data[:,col], label=headers[col])
    
    
    
plt.legend()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
```

```python
for col in range(1,num_columns):
    plt.figure()  # creates a separate graph for every loop
    #print(the column number is {col}.)
    
    plt.plot(data[:,col], label=headers[col])
    plt.legend()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstroms)')
    plt.savefig('all_samples.png', dpi=200)

```

```python
plt.plot(data[:,0], data[:,1])
```
